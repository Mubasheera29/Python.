```python
score={"Jane":90, "Bill":80,"Elon":85,"Tom":75,"Tim":95}
names=pd.Series(score) # Convert to Series
names
```

```python
names["Tim"]
```

```python
names[names>=85]
```

```python
names["Tom"]=60
names
```

```python
names[names<=80]=83
names
```
